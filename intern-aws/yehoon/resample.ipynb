## Resample

```python
df_resam = df.resample(rule='15s').first()
# np.round(df_resam.head(), 2)
```

```python
df_resam.errorCode.value_counts()
```

```python
df_mean = df_resam.fillna(method='ffill')
```

```python
df_mean
```

```python
df_mean.to_pickle('../data/resample.pkl')
```

```python
x_data = df_mean.iloc[:, 1:].as_matrix()
y_data = df_mean.iloc[:, 0].as_matrix().reshape(-1, 1)
y_data.shape, x_data.shape
```

```python
y_data
```

```python
x_data = x_data.astype('int')
y_data = y_data.astype('int')
x_data, y_data
```

## Decision Tree

```python
from sklearn import preprocessing # Min-Max Standardzation

min_max_scaler = preprocessing.MinMaxScaler()
x_data = min_max_scaler.fit_transform(x_data)

import numpy as np 

training_idx = np.random.randint(y_data.shape[0], size=int(y_data.shape[0] * 0.8))
test_idx = np.random.randint(y_data.shape[0], size=int(y_data.shape[0] * 0.2))

x_train, x_test = x_data[training_idx,:], x_data[test_idx,:]
y_train, y_test = y_data[training_idx,:], y_data[test_idx,:]

x_train.shape, x_test.shape
```

```python
from sklearn import tree
from sklearn import metrics
qoe_tree = tree.DecisionTreeClassifier(criterion='gini', splitter='random', max_depth=5)
qoe_tree.fit(x_train, y_train)
```

```python
y_pred_tr =qoe_tree.predict(x_test)
print('Accuracy: %.2f' % metrics.accuracy_score(y_test, y_pred_tr))
```

```python
print(metrics.classification_report(y_test, y_pred_tr))
```
