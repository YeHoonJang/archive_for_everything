## Label Encoding - Device

```python
df['device'].value_counts()
```

```python
le_divice = preprocessing.LabelEncoder()
df['device'] = le_divice.fit_transform(df['device'].astype(str))
le_divice.classes_
```

## Label Encoding - NetworkType

```python
df['networkInfo.type'].value_counts()
```

```python
le_network = preprocessing.LabelEncoder()
df['networkInfo.type'] = le_network.fit_transform(df['networkInfo.type'].astype(str))
le_network.classes_
```

```python
df.shape
```

## Label Encoding - DRM system

```python
le_drm = preprocessing.LabelEncoder()
df['licenseInfo.drmSystem'] = le_drm.fit_transform(df['licenseInfo.drmSystem'].astype(str))
le_drm.classes_
```

```python
df['licenseInfo.drmSystem'].value_counts()
```

## Counting Data - FrameDropped

```python
for row, val in enumerate(df['frameDropped']):
    if isinstance(val, float):
        df['frameDropped'][row] = 0
    else:
        df['frameDropped'][row] = len(val)
```

```python
df['frameDropped'].value_counts()
```

## Counting Data - QualityChanged

```python
df['qualityChangedCount'] = None
```

```python
for row, val in enumerate(df['qualityChanged']):
    if isinstance(val, float):
        df['qualityChangedCount'][row] = 0
    else:
        df['qualityChangedCount'][row] = len(val)
```

```python
df['qualityChangedCount'].value_counts()
```

```python
df
```

## ContentUrl -> Content type

```python
df['content_type'] = ''
```

```python
for i in range(0,len(df)):
    try: 
        if 'video' in df['contentUrl'][i].split('/'):
            df['content_type'][i] = 'video'
        elif 'audio' in df['contentUrl'][i].split('/'):
            df['content_type'][i] = 'audio'
        else : 
            df['content_type'][i] = df['contentUrl'][i].split('/')[6]
    except:
        df['content_type'][i] = 'nothing'
        pass
```

```python
df = df[df['content_type'] != 'nothing']
```

```python
le_content = preprocessing.LabelEncoder()
df['content_type'] = le_content.fit_transform(df['content_type'].astype(str))
le_content.classes_
```

```python
df['content_type'].value_counts()
```

```python
del df['contentUrl']
```

## bufferings -> bufferingTime ( total buffering time)

```python
df['bufferingTime']= 0
```

```python
for row, val in enumerate(df['bufferings']):
    if isinstance(val, float):
        df['bufferingTime'].iloc[row] = 0
    else:
        if len(df.bufferings.iloc[row])!=0:
            for i in range(len(df.bufferings.iloc[row])):
                df['bufferingTime'].iloc[row] += df.iloc[row]['bufferings'][i]['duration']
        else:
            df['bufferingTime'].iloc[row]= 0
```

```python
del df['bufferings']
```

```python
del df['qualityChanged']
```

```python
del df['errorString']
```
