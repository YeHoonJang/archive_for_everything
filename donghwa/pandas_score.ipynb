## average 는 전과목 평균 / 3과목 평균을 구해서 추가

#### 평균 = (수학+영어+국어)/3

```python
math = score.math
```

```python
type(math)
```

```python
english = score.english
```

```python
korean = score.korean
```

```python
total = math + english + korean
total
```

```python
average_3 = total/3
average_3
```

```python
score['average_3'] = average_3
```

```python
score['avg_gap'] = score['average'] - score['average_3']
```
