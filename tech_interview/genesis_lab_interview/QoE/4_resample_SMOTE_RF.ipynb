```python
from collections import Counter
from imblearn.over_sampling import SMOTE
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
import pandas as pd
import numpy as np
```

```python
df = pd.read_pickle('../data/resample.pkl')
df
```

```python
df['errorCode'].value_counts()
```

```python
df = df[df.errorCode != -2147483648.0]
df = df[df.errorCode != -12.0]
df = df[df.errorCode != -100874.0]
df = df[df.errorCode != -100870.0]
df = df[df.errorCode != -100859.0]

df.errorCode.value_counts()
```

```python
x = df.loc[:, df.columns!='errorCode'].astype('float32')
y = df['errorCode'].astype('float32')
x.shape, y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=50)
x_train.shape, x_test.shape, y_train.shape, y_test.shape 
```

```python
print('Original dataset shape {}'.format(Counter(y_train)))
sm = SMOTE(random_state=42)
x_res , y_res = sm.fit_sample(x_train, y_train)
print('Resampled dataset shape {}'.format(Counter(y_res)))
```

### Decision Tree (x)

```python
from sklearn import tree
from sklearn import metrics
qoe_tree = tree.DecisionTreeClassifier(criterion='gini', splitter='random', max_depth=5)
qoe_tree.fit(x_res, y_res)
```

```python
y_pred_tr =qoe_tree.predict(x_test)
print('Accuracy: %.2f' % metrics.accuracy_score(y_test, y_pred_tr))
```

```python
print(metrics.classification_report(y_test, y_pred_tr))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
qoe_forest = RandomForestClassifier()
qoe_forest.fit(x_res, y_res)
```

```python
y_pred_tr = qoe_forest.predict(x_test)
print('Accuracy: %.2f' % metrics.accuracy_score(y_test, y_pred_tr))
```

```python
print(metrics.classification_report(y_test, y_pred_tr))
```
