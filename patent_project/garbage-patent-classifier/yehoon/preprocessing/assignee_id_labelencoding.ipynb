```python
assignee_list = []
for i in patent_id_list:
    
    sql = "select * from uspto.patent_assignee where patent_id={}".format(i)
    curs.execute(sql)
    rows = curs.fetchall()
    print(rows)
    assignee_list.append(rows)
```
