### 라이브러리 불러오기

```python
# 그래프 라이브러리
import matplotlib.pyplot as plt
%matplotlib inline

# word cloud 라이브러리
from wordcloud import WordCloud, STOPWORDS

# 행렬, 다차원 데이터 지원 라이브러리
import numpy as np

# 파이썬 이미징 라이브러리
from PIL import Image
```

### 앨리스 텍스트 불러오기

```python
txt = open('alice.txt').read()
```

```python
txt
```

### 앨리스 그림 불러온 후 행렬화

```python
alice_mask = np.array(Image.open("alice_mask.png"))
```

```python
alice_mask
```

```python
Image.open("alice_mask.png")
```

### 좌표평면에 앨리스 그림 그리기

```python
plt.figure(figsize=(8,8))
plt.imshow(alice_mask,cmap=plt.cm.gray,interpolation='bilinear')
plt.show()
```

```python
alice_wc = WordCloud(background_color="white", max_words=2000, mask=alice_mask)
alice_wc = alice_wc.generate(txt)
alice_wc.words_
```

```python
plt.figure(figsize=(12,12))
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

### 불필요한 단어 제거

```python
stopword = set(STOPWORDS)
stopword.add('said')
```

```python
alice_wc = WordCloud(background_color="white",
               max_words=2000, mask=alice_mask, stopwords=stopword)
alice_wc = alice_wc.generate(txt)
alice_wc.words_
```

```python
plt.figure(figsize=(12,12))
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
stopword.add('Alice')
```

```python
alice_wc = WordCloud(background_color="white",
               max_words=2000, mask=alice_mask, stopwords=stopword)
alice_wc = alice_wc.generate(txt)
alice_wc.words_
```

```python
plt.figure(figsize=(12,12))
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
sw_stopwd = set(STOPWORDS)
sw_stopwd.add("int")
sw_stopwd.add("ext")
```

```python
sw_wc = WordCloud(background_color="black", max_words=2000, mask=sw_mask, 
                 stopwords=sw_stopwd)
sw_wc = sw_wc.generate(sw_txt)
sw_wc.words_
```

```python
import random
def grey_color_func(word, font_size, position, orientation, random_state=None,
                    **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60, 100)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(sw_wc.recolor(color_func=grey_color_func, random_state=3), interpolation='bilinear')
plt.axis("off")
plt.show()
```
