```python
# 판다스 라이브러리 호출
import pandas as pd
```

```python
# csv 데이터 읽어오기
df = pd.read_csv("성적 예측 분석 데이터 수집 설문조사.csv")
df.head() # 5개의 데이터만 보여주는 함수 
# head(10) 하면 10개의 데이터 출력
```

```python
# 데이터 프레임의 컬럼 확인
df.columns
```

```python
#영어, 수학, 국어, 전과목 평균 점수만을 추출해서 score라는 데이터 프레임 생성
score = df[['영어(영어, 영어 I, 심화 영어 독해 I)', 
          '수학(수학, 수학 I, 수학연습 I, 수학연습 II, 확률과 통계, 기하와 벡터)',
         '국어 (국어, 문학, 고전)', '19-1 전과목 중간고사 성적 평균']]
```

```python
#score 확인
score
```

```python
# 컬럼의 이름을 바꿔주기 위해 colums_name 리스트 데이터 생성
# 리스트 데이터 순서 유의!
colums_name = ["english", 'math', 'korean', 'average']
```

```python
#score 데이터 프레임의 컬럼 이름을 위 영어 이름으로 변경
score.columns = colums_name
```

```python
#확인
score.head()
```

## average 는 전과목 평균 / 3과목 평균을 구해서 추가

#### 평균 = (수학+영어+국어)/3

```python
# 데이터 간 연산을 위해 score 데이터 프레임에서 math 데이터만 따로 추출
math = score.math
```

```python
# 타입 확인
type(math)
```

```python
# 데이터 확인
math.head()
```

```python
# 영어 점수 추출
english = score.english
```

```python
# 국어 점수 추출
korean = score.korean
```

```python
# total 변수 생성 후 세 과목을 더함
# 파이썬 리스트처럼 데이터들이 합집합 되는 것이 아닌 진짜 덧셈 연산이 된 값이 나타남
# 판다스에서는 시리즈 데이터 덧셈 연산을 하면 같은 인덱스 데이터끼리 더해짐
# ex) 0번째 세 과목 점수 합 = (영어 점수 + 수학 점수 + 국어 점수) = 100 + 77.4 + 95 = 272.4
# total 시리즈 데이터 0번째 확인 결과 272.4
total = math + english + korean
total
```

# 세 과목의 평균 시리즈 데이터를 average_3이라는 변수에 할당
# 

```python
average_3 = total/3
average_3
```

```python
# score 데이터 프레임에 'average_3' 라는 컬럼을 생성하고 average_3 시리즈 데이터 할당
score['average_3'] = average_3
```

```python
# 확인
score.head()
```

```python
# 전 과목 평균과 세 과목 평균의 차를 구하고 싶음
# 위 처럼 데이터 프레임에서 시리즈 데이터를 추출해서 연산 후 다시 데이터 프레임에 추가 하는 게 아닌
# 데이터 프레임 상에서 바로 작업

# socre 데이터 프레임에 'avg_gap' 컬럼은 생성 후
# 전과목 평균 시리즈 데이터와 세 과목 시리즈 데이터의 차를 바로 할당
score['avg_gap'] = score['average'] - score['average_3']
```

```python
# 확인
score
```
